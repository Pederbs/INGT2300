```python
import numpy as np
from IPython.display import Image
Image(filename='Ass1a.jpg')
```

```python
num_a = (0.1-0.09)/0.09 # prosessen fr 0.1kg metall produserer 0.1/0.09 kWh og gir bort 1kw str da igjenn med 0.11... samme som str der
num_b = 0.25
num_c = 1.4*10
num_d = 1
Image(filename='Ass1b.jpg')
```

```python
A = np.array([[0.1+num_d, -0.1, -num_d], 
             [-1, 1+num_a, -num_a], 
             [0, 0, 1]])
Image(filename='Ass1c.jpg')
```

```python
B = np.array([[-10, 0, 0], 
              [num_c, 0.5, num_b]])
Image(filename='Ass1d.jpg')
```

```python
f = np.array([[0], [0], [1]])
r = B@np.linalg.inv(A)@f
#print(r)
co2_dingser = r[1, 0]
print(co2_dingser)
```
