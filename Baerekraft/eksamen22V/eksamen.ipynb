```python
n1 = 98000
n2 = 140000
n13 = 14000
n21 = 3000
n12 = n1
n23 = n2


A = np.array([[n1, -n12, -n13],
              [-n21, n2, -n23],
              [0, 0, 25]])
Image(filename='opg1b.jpg')
```

```python
w1 = 14700
q1 = 93
p1 = 4900
w2 = 0
q2 = 255
p2 = 13400
w3 = 2800
q3 = 285
p3 = 15000

B = np.array([[-w1, -w2, -w3],
              [q1, q2, q3],
              [p1, p2, p3]])
Image(filename='opg1c.jpg')
```

```python
f = np.array([[0], [0], [1]])
r = B@np.linalg.inv(A)@f
nox_vindmolle = r[1,0]
print(nox_vindmolle)
```
